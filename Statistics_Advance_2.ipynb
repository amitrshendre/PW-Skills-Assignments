Q1. What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to make decisions about the properties of a population based on a sample. It involves formulating a null hypothesis (H0) and an alternative hypothesis (H1), then using sample data to determine whether to reject H0.
- **Example:** Testing whether a new drug is more effective than the current standard treatment.
-----------------------------------------------------------------
Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis (H0) is a statement that there is no effect or no difference, and it serves as the default assumption. The alternative hypothesis (H1) is a statement that there is an effect or a difference. Hypothesis testing aims to determine whether there is enough evidence to reject H0 in favor of H1.
- **Example:** H0: The new drug has no effect. H1: The new drug has an effect.
-----------------------------------------------------------------
Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
- The significance level (alpha) is the probability of rejecting the null hypothesis when it is true. It is a threshold set by the researcher, commonly 0.05 or 0.01, to determine the strength of the evidence against H0. It is important because it controls the likelihood of making a Type I error.
- **Example:** Setting alpha to 0.05 means there is a 5% chance of rejecting H0 when it is true.
-----------------------------------------------------------------
Q4. What are Type I and Type II errors? Give examples of each.
- A Type 1 error occurs when the null hypothesis is rejected when it is true (false positive). A Type 2 error occurs when the null hypothesis is not rejected when it is false (false negative). The significance level (alpha) controls the probability of a Type 1 error, while the power of the test affects the probability of a Type 2 error.
- **Example:** Type 1 error: Concluding a new drug is effective when it is not. Type 2 error: Concluding a new drug is not effective when it is.
-----------------------------------------------------------------
Q5. What is the difference between a Z-test and a T-test? Explain when to use
each.
A Z-test is used when the population standard deviation is known and the sample size is large (typically 30 or more), while a T-test is used when the population standard deviation is unknown and the sample size is small (typically less than 30).
**When to Use a Z-test**
- Known Population Standard Deviation: The primary condition is that you know the population standard deviation (Ïƒ).
- Large Sample Size: Your sample size should be large, generally 30 data points or more.
**When to Use a T-test**
- Unknown Population Standard Deviation: The population standard deviation is unknown, so you must estimate it from the sample data.
- Small Sample Size: You are working with a small sample, typically fewer than 30 data points.
- Data Distribution: The data sets should follow a normal or approximately normal distribution.

```python
# Practical question
'''
Q6. : Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

'''
'''
Answer:-6
'''
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5 # Probability of success

# Generate random samples from the binomial distribution
# 'size' determines how many random samples to generate
num_samples = 10000
data = np.random.binomial(n, p, size=num_samples)

# Create a histogram of the generated data
plt.hist(data, bins=np.arange(-0.5, n + 1.5, 1), edgecolor='black', alpha=0.7)

# Add labels and title to the plot
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')

# Set x-axis ticks to represent whole numbers of successes
plt.xticks(np.arange(0, n + 1, 1))

# Display the plot
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Practical question
'''
Q7. Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
'''
'''
Answer:-7
'''
import numpy as np
from statsmodels.stats.weightstats import ztest

# Define the sample data provided
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Assume population parameters for the Z-test
pop_mean = 50.0  # Hypothesized population mean (Î¼0)
pop_std = 1.5   # Known population standard deviation (Ïƒ)

# Convert the sample data to a NumPy array
sample_data_np = np.array(sample_data)

# Perform the one-sample Z-test
z_statistic, p_value = ztest(x1=sample_data_np, value=pop_mean, alternative='two-sided')

# Significance level
alpha = 0.05

# Print the results
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (alpha): {alpha}")
print("-" * 30)

# Interpret the results
if p_value < alpha:
    print("Interpretation: Reject the null hypothesis.")
else:
    print("Interpretation: Fail to reject the null hypothesis.")
```

```python
# Practical question
'''
Q8. Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
'''
'''
Answer:-8
'''
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 1. Simulate data from a normal distribution
mu_true = 10  # True mean of the population
sigma_true = 2  # True standard deviation of the population
sample_size = 100
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=mu_true, scale=sigma_true, size=sample_size)

# 2. Calculate the 95% confidence interval for the mean
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
standard_error = sample_std / np.sqrt(sample_size)

# For a 95% confidence interval, alpha = 0.05, so alpha/2 = 0.025
# We use the t-distribution for confidence intervals when the population standard deviation is unknown.
degrees_freedom = sample_size - 1
t_critical = stats.t.ppf(0.975, degrees_freedom) # For a two-tailed 95% CI

margin_of_error = t_critical * standard_error
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: [{lower_bound:.2f}, {upper_bound:.2f}]")

# 3. Plot the data using Matplotlib
plt.figure(figsize=(10, 6))
plt.hist(data, bins=15, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Simulated Data Histogram')

# Plot the theoretical normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mu_true, sigma_true)
plt.plot(x, p, 'k', linewidth=2, label='True Normal Distribution')

# Plot the sample mean and confidence interval
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=1.5, label=f'Sample Mean ({sample_mean:.2f})')
plt.axvline(lower_bound, color='green', linestyle='dotted', linewidth=1.5, label='95% CI Lower Bound')
plt.axvline(upper_bound, color='green', linestyle='dotted', linewidth=1.5, label='95% CI Upper Bound')

plt.title('Simulated Normal Distribution Data with 95% Confidence Interval')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Practical question
'''
Q9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
'''
'''
Answer:-9
'''
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_visualize_zscores(data):
    """
    Calculates the Z-scores for a dataset and visualizes the standardized data
    using a histogram.

    Args:
        data (array-like): A list or NumPy array of numerical data.

    Returns:
        numpy.ndarray: An array containing the calculated Z-scores.
    """
    # Convert data to a NumPy array to ensure compatibility
    data = np.array(data)

    # Handle the edge case of no data
    if data.size == 0:
        print("Error: The dataset is empty.")
        return np.array([])

    # Calculate the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Handle the edge case of zero standard deviation
    if std_dev == 0:
        print("Warning: Standard deviation is zero. Z-scores cannot be calculated.")
        return np.zeros_like(data)

    # Calculate the Z-scores
    z_scores = (data - mean) / std_dev

    # Visualize the standardized data using a histogram
    plt.figure(figsize=(10, 6))
    plt.hist(z_scores, bins='auto', edgecolor='black', color='skyblue')
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-Score (Number of Standard Deviations from the Mean)')
    plt.ylabel('Frequency')
    plt.axvline(0, color='red', linestyle='dashed', linewidth=1, label='Mean (Z=0)')
    plt.grid(axis='y', alpha=0.75)
    plt.legend()
    plt.show()

    return z_scores

# Example usage with a sample dataset
sample_data = [65, 70, 72, 80, 85, 88, 90, 95, 100, 110, 125]

# Call the function with the sample data
z_scores = calculate_and_visualize_zscores(sample_data)

# Print the original data and calculated Z-scores
print("Original Data:", sample_data)
print("Calculated Z-Scores:", np.round(z_scores, 2))
```
